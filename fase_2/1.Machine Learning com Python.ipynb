## Scikit-Learn

```python
df_diabetes = pd.read_csv('../dados/00.fase 2/diabetes.csv')
df_diabetes.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_diabetes.drop(["Class variable"], axis=1)
x.head(3)
```

```python
y = df_diabetes['Class variable'] # 1 = tem diabetes, 2 = não tem; é o alvo que queremos atingir com o nosso modelo
y.head(3)
```

### Separando os dados

Treinaremos o algoritmo em ~70-80% da base, e deixaremos o resto para "teste", para ver se o nosso algoritmo consegue fazer previsões corretas.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

### KNeighborsClassifier
Temos dados, e temos uma "resposta" para esses dados (diabético sim ou não), então vamos utilizar um algoritmo que trabalha bem com classificações.


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
#n_neighbors = quantos vizinhos próximos (outros dados parecidos) ele vai olhar
```

```python
knn.fit(x_train, y_train)
```

```python
accuracy = knn.score(x_test, y_test)
print(f'Acertou {accuracy*100:f}% dos testes.')
```

Dependendo da base, 70% pode ser **alto** ou **baixo**.

## Teste Shapiro
Que tipo de distribuição é essa? (Normal, assimétrica, etc)

```python
from scipy.stats import shapiro
```

```python
def distribution_normality (df):
    stat, p = shapiro(df)

    print(f'estatística de teste: {stat:.4f}\np: {p:.10f}')

    if p > 0.05:
        print('Não há evidência o suficiente para rejeitar a hipótese de normalidade\n')
    else:
        print('A hipótese de normalidade é rejeitada\n')
```

```python
distribution_normality(df_rs_studio_loudness)
distribution_normality(df_rs_live_loudness)
```

Entendemos que nossos dados não seguem uma distribuição normal, então não podemos realizar testes de hipóteses que seguem esse princípio, também chamados de testes não paramétricos.

## Teste Mannwhitneyu
Ver a semelhança / diferença entre duas estatísticas.

O teste de Mann-Whitney U é um teste estatístico que compara duas amostras para determinar se elas têm distribuições diferentes. Ele classifica as observações de cada amostra(ou seja ordená-las em uma determinada ordem com base nos valores que representam), soma as classificações em cada amostra e compara as somas para obter um valor U. Se esse valor for grande o suficiente, podemos concluir que as médias das duas amostras são significativamente diferentes. Em resumo, o teste de Mann-Whitney U é usado para verificar se duas amostras são estatisticamente diferentes entre si, mesmo que elas não sigam uma distribuição normal.

```python
from scipy.stats import mannwhitneyu
```

```python
def mannwhit(df1, df2):
    stats, p = mannwhitneyu(df1.sample(len(df1)), df2.sample(len(df2)), alternative="less")

    print(f'estatística de teste: {stats:.4f}\np: {p:.10f}')

    alpha = 0.05
    if p<alpha:
        print('Diferença estatisticamente significante')
    else:
        print('Diferença estatisticamente insignificante')
```

```python
mannwhit(df_rs_studio_loudness, df_rs_live_loudness)
```

Conseguimos um p-valor menor do que 0.05, ou seja, podemos confirmar que os albuns gravados em studio, de fato, são menos barulhentos do que quando são ao-vivo.
