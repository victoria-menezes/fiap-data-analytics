Fazendo uma função para não precisarmos reescrever o mesmo código várias vezes

```python
def plota_gastos_por_mes(df : pd.DataFrame, figsize : tuple = (10,6), lim : tuple = (0,600), anchor : tuple = (1.35, 0.5), colors : list = ['c', 'm', 'b', 'k']):
    # Creating a color Cycler
    color_cycle = cycler(color=colors)
    
    axis = df.T.plot(figsize=figsize)
    plt.rc('axes', prop_cycle=color_cycle)    
    plt.legend(loc='right', bbox_to_anchor=anchor)
    axis.set_ylabel(atendimento_ylabel) 
    axis.set_xlabel(atendimento_xlabel)
    axis.set_ylim(lim[0],lim[1])
```

exemplo de como usar:

```python
np.random.seed(524387)
plota_gastos_por_mes(df_gastos_preenchido[meses_filtrados].sample(n=3), lim=(0,150))
```

```python
np.random.seed(524387)
plota_gastos_por_mes(df_gastos_preenchido[meses_filtrados].sample(n=3), lim=(0,150))
```

Cuidado ao alterar os limites, pois ela pode acabar criando relações errôneas, especialmente ao aumentar o limite mínimo do gráfico.

```python
plota_gastos_por_mes(df_gastos_numericos_sem_total[df_gastos_numericos_sem_total.columns[-12:]].head(4), lim=(0,500)) # ultimos 12 meses
```

```python
plota_gastos_por_mes(df_gastos_numericos_sem_total[df_gastos_numericos_sem_total.columns[-12:]].head(4), lim=(80,500)) # ultimos 12 meses
```

```python
plota_gastos_por_mes(df_gastos_numericos_sem_total[df_gastos_numericos_sem_total.columns[-12:]].head(3)) # ultimos 12 meses
```

```python
plota_gastos_por_mes(df_gastos_numericos_sem_total[df_gastos_numericos_sem_total.columns].head(3), figsize=(6,6), anchor=(1.5,0.5)) # ultimos 12 meses
plota_gastos_por_mes(df_gastos_numericos_sem_total[df_gastos_numericos_sem_total.columns].head(3), figsize=(12,3), anchor=(1.25,0.5))
```

Perceba que até mesmo o tamanho do gráfico altera nossa percepção dos dados.

## Desafios
- Atualizar esse gráfico acima para refletir o SEU estado; incluir grid, eixos, etc
- Colorir o estado de referência com uma cor diferente; colorir os outros estados de acordo com gasto maior ou menor

```python
def maior_menor(val, reference_val):
    return 'cyan' if float(val) > float(reference_val) else 'magenta' if float(val) < float(reference_val) else 'blue'
```
