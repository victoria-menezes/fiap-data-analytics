```python
def get_contents(elem : Tag) -> str:
    '''
        Returns the content of the bs4 element as a string.

        Parameters
        ----------
        elem : bs4.element.Tag        

    '''
    return str(elem.decode_contents()).strip()
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_population'
req = requests.get(url)
soup = BeautifulSoup(req.content, "html.parser")
table_class = 'references-small'
table = soup.find("table", attrs={'class':'sortable'})
df_countries_eu = pd.read_html(url, index_col=1)[1]
df_countries_eu = df_countries_eu.drop(['Unnamed: 0','% change','Official figure', 'Official date', 'Unnamed: 6'], axis=1).reset_index()
df_countries_eu = df_countries_eu.drop(0)
df_countries_eu.columns = ['pais_eng','populacao']
df_countries_eu['continente'] = 'Europa'


df_countries_eu['pais'] = df_countries_eu['pais_eng'].map(countries_translation)
df_countries_eu = df_countries_eu[['continente','pais_eng', 'pais', 'populacao']]
df_countries_eu.head(5)
```
